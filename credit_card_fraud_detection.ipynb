**CLASS DISTRIBUTION**

```python
#distribution of class
display(df['Class'].value_counts())

print('Non-fraud is ', round(df['Class'].value_counts()[0]/len(df) * 100, 2), '% of the dataset')
print('Fraud is ', round(df['Class'].value_counts()[1]/len(df) * 100, 2), '% of the dataset')

#plot the visualization
plt.figure(figsize=(8, 5))
sns.countplot(data = df, x = 'Class', hue='Class')
plt.title("Class Distribution")
plt.xlabel("Class (0: Non-Fraud, 1: Fraud)")
plt.ylabel("Count")
plt.show()
```

* There is a significant imbalance in the dataset with the majority of transactions being non-fraudulent (class 0) and only 0.17% being fraudulent (class 1). This will result in a biased model which will perform well on the majority class but fail to detect fraud transactions.

* It is important to balance the data to ensure that the model is able to accurately detect fraud

**DISTRIBUTION OF TIME**

```python
#visualize the distribution of time
plt.figure(figsize=(12, 6))
sns.histplot(df[df.Class == 1]['Time'], bins=50, kde=True, color='red', stat='density', alpha=0.5, label='Fraud') #distribution of fraud transactions
sns.histplot(df[df.Class == 0]['Time'], bins=50, kde=True, color='blue', stat='density', alpha=0.2, label='Non-Fraud')# distribution of regular transactions

plt.title('Distribution of Time Variable by Transaction Class')
plt.xlabel('Time (seconds)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

* There is a significant overlap between the two which indicates that the Time variable might not effectively differentiate between fraudulent and non-fraudulent transactions. So we will drop it.

```python
df = df.drop(['Time'], axis=1 )
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='Amount', data=df, hue='Class', palette=['green', 'red'])
plt.title('Transaction Amounts by Class')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Transaction Amount')
plt.ylim(0, df['Amount'].quantile(0.95))
plt.show()
```

```python
# scaling the amount column
scaler = StandardScaler()
df['Amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))
df.head()
```

```python
#set features and target
X = df.drop(columns=['Class'])
y = df['Class']
```

```python
#split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

**SMOTE**

```python
#apply SMOTE to handle class imbalance
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

print("Class distribution after SMOTE:", y_train_resampled.value_counts())
```

**MODEL 1 : Random Forest**

```python
#fit model
rf_model= RandomForestClassifier(random_state=42)
rf_model.fit(X_train_resampled, y_train_resampled)
```

```python
#evaluate the model
y_pred = rf_model.predict(X_test)
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
#confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 5))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=["Non-Fraud", "Fraud"], yticklabels=["Non-Fraud", "Fraud"])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
#calculate ROC-AUC
roc_auc = roc_auc_score(y_test, rf_model.predict_proba(X_test)[:, 1])
print("ROC-AUC Score:", roc_auc)
```
